# SOAL 4

Diberikan: x = [1, 2, 3, 4, 5], y = [3, 5, 7, 9, 11]
Hitung y = mx + c, tampilkan nilai m dan c, dan visualisasikan dalam 1 plot.

````python
x = [1, 2, 3, 4, 5]
y = [3, 5, 7, 9, 11]

x = np.array(x)
y = np.array(y)

A = np.vstack([x, np.ones(len(x))]).T
Y = y[:, np.newaxis]

alpha = np.linalg.inv(np.dot(A.T,A))
alpha = np.dot(alpha, A.T)
alpha = np.dot(alpha, Y)

m, c = alpha[0,0], alpha[1,0]

plt.plot(x, y, "ro")
plt.plot(x, m * x + c, "-b")
plt.show()
````

# C. NEWTON RAPHSON

# SOAL 7

Gunakan fungsi
```yaml
f(x): x3 - x - 2
```
tebakan awal = 1.5
tolerance = 0.01
max iteration = 20

````python
def f(x):
    return (x ** 3) - x - 2

def g(x):
    return 3 * (x ** 2) - 1

def newton_raphson(x0, tolerance = 0.01, iter = 20, cur_iter = 1):
    x1 = x0 - f(x0) / g(x0)
    fx1 = f(x1)

    print(f"Iteration {cur_iter} new root = {x1}")

    if (np.abs(fx1) < tolerance):
        print(f"Root is found at {cur_iter} iteration = {x1}")
        return
    else:
        newton_raphson(x1, tolerance, iter, cur_iter + 1)
newton_raphson(1.5)
````

# RIEMANN INTEGRATION

# SOAL 10

Hitung pendekatan
$$
\int_{1}^{5}\bigl(x^2 + 2x + 1\bigr)\,dx
$$

dengan 1000 titik, metode right dan middle Riemann

````python
def f(x):
    return (x ** 2) + 2 * x + 1

a = 1
b = 5
n = 1000

width = (b - a) / (n - 1)

x = np.linspace(a, b, n)
y = f(x)

x_mid = (x[:-1] + x[1:]) / 2
y_mid = f(x_mid)

mid_riemann = width * np.sum(y_mid)
right_riemann = width * np.sum(y[1:])

print(f"Right: {right_riemann}")
print(f"Middle: {mid_riemann}")
````

# SOAL 11

$$

dengan 1000 titik, metode right dan middle Riemann

````python
def f(x):
    return np.sin(x) + (x ** 2)

def g(x):
    return np.cos(x) + 2 * x

a = 0
b = 3.14
n = 5000

width = (b - a) / (n - 1)

x = np.linspace(a, b, n)
y = f(x)

x_mid = (x[:-1] + x[1:]) / 2
y_mid = f(x_mid)

mid_riemann = width * np.sum(y_mid)
right_riemann = width * np.sum(y[1:])
left_riemann = width * np.sum(y[:-1])

print(f"Mid: {mid_riemann}")
print(f"Right: {right_riemann}")
print(f"Left: {left_riemann}")
````
